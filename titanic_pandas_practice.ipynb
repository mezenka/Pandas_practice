```python
# refresh the dataframe to get read of all previous changes
titanic = pd.read_csv("https://raw.githubusercontent.com/mezenka/Pandas_Titanic/main/titanic_train.csv")
titanic
```

```python
# concat
data_to_add={'survived':[1,0],'pclass':[1,3],'name':['Miss. Rose Dewitt Bukater','Mr. Jack Dawson'],'sex':['female','male'],'age':[17,20],'sibsp':[2,0],'parch':[0,0],'fare':[49.5042,19.2583]}
addendum = pd.DataFrame(columns = titanic.columns, data=data_to_add)
addendum
```

```python
pd.concat([titanic,addendum])
```

```python
# save into new dataframe, refresh index
titanic_c = pd.concat([titanic,addendum]).reset_index(drop=True)
titanic_c
```

```python
# merge
data_to_add = {'pclass':[1,2,3],'total_cabin_qty':[371,105,84]}
extra_info = pd.DataFrame(columns=['pclass','total_cabin_qty'], data=data_to_add)
extra_info
```

```python
titanic.merge(extra_info, how='left', on='pclass')
```

**Aggregations**

```python
# describe
titanic.describe()
```

```python
titanic.mean()
```

```python
titanic['age'].mean(),titanic['pclass'].mean(),titanic['fare'].mean(),titanic['survived'].mean() # statistics for several variables
```

```python
# value_counts()
titanic_srv = titanic[titanic['survived'] == 1].reset_index()
titanic_srv['age'].value_counts(ascending = False) # counting numbers of the survived by age category
```

```python
titanic_srv['sex'].value_counts(ascending = False) # counting numbers by sex of those survived
```

```python
titanic_nsrv = titanic[titanic['survived'] == 0].reset_index()
len(titanic_nsrv) # number of those who did not survive
```

```python
titanic_nsrv['sex'].value_counts(ascending = False) # counting numbers by sex of those who did not survive
```

```python
# groupby
titanic.groupby('sex')['age'].mean() # average ages by sex
```

```python
titanic.groupby('sex')['passenger_id'].count() # numbers by sex
```
