```python
import re
```

```python
quote = "There's only one thing I have more than lying: skim milk. Which is whater that's lying about being milk. - Ron Swanson"

re.search("milk", quote)
```

```python
re.search("lying", quote)
```

```python
re.search("milk", quote).group()
```

```python
re.findall("milk", quote)
```

```python
# to count number of appearances
len(re.findall("milk", quote))
```

```python
re.split("milk", quote)
```

```python
re.split("\.", quote) # splitting text into sentences using periods'.'
```

```python
re.sub("I", "you", quote)
```

```python
re.sub("milk", "dairy", quote)
```

```python
re.sub("milk", "dairy", quote, count=2)
```

```python
string = 'I like the mountains in the spring. 24687'

re.findall('[a-zA-Z0-9]',string)
```

```python
# any character
string = 'I have 123,456 koalas!'

re.findall('[0-4]',string)
```

```python
string = 'you can see sea shells by the sea shore'

re.findall('.',string)
```

```python
# any character between specified letters
string = 'You can see sea shells by the sea shore'

re.findall('s.a',string)
```

```python
# only {1} any character between the specified letters
string = 'You can see sea shells by the sea shore. sba'

re.findall('s.{1}a',string)
```

```python
string = "Well well well... If it isn't Will Wilmer "

re.findall('W.{2}l',string)
```

```python
# ^-beginning of string
# $-end of string
string = "Happy birthday to you. Happy birthday to you. Happy birthday der Alex, happy birthday to you."

re.findall('^Happy',string)
```

```python
# ^-beginning of string
# $-end of string
string = "Happy birthday to you. Happy birthday to you. Happy birthday der Alex, happy birthday to you."

re.findall('^you',string)
```

```python
# ^-beginning of string
# $-end of string
string = "Happy birthday to you. Happy birthday to you. Happy birthday der Alex, happy birthday to you."

re.findall('you$',string)
```

```python
# ^-beginning of string
# $-end of string
string = "Happy birthday to you. Happy birthday to you. Happy birthday der Alex, happy birthday to you."

re.findall('you.$',string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble has Thrice hurt me'
re.findall('Thi.*s', string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble ha Thrice hurt me'
re.findall('Thi.+s', string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble ha Thrice hurt me'
re.findall('Thi.?s', string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble has Thrice hurt me'
re.findall('Thi.*e', string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble has Thrice hurt me'
re.findall('Thi.+e', string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble has Thrice hurt me'
re.findall('Thi.?e', string)
```

```python
# * zero or more
# + one or more
# ? zero or one
string = 'This Thing called a Thimble has Thrice hurt me'
re.findall('Thi.{3}?e', string)
```

```python
string = 'I hate and I love balloon animals. They are beautiful.'

re.findall('hate|beautiful', string)
```

```python
string = 'I hate and I love balloon animals. They are beautiful.'

re.findall('lovely|beautiful', string)
```

```python
string = 'I like cats. You like cats. We all like cats.'

re.findall('.', string)
```

```python
# looking for spepcific values with '\'
string = 'I like cats. You like cats. We all like cats.'

re.findall('\.', string) # '.' no longer ANY caracter, with '\' it actually becomes the '.'
```

```python
string = 'I like cats. You like cats. We all like cats.'

re.findall('\?', string)
```
