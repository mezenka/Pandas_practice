Given a pandas DataFrame `df` with columns
'A', 'B', and 'C', filter rows where column 'A' is
greater than 10, and select only columns 'A'
and 'C' for the resulting subset.

```python
# Sample DataFrame
data = {'A':[15,8,20,12],
        'B':[10,5,7,22],
        'C':[25,18,30,22]}
df = pd.DataFrame(data)

# Solution
result = df[df['A']>10][['A','C']]
result
```

Given a DataFrame `df` containing sales data with columns 'Product', 'Quantity', and 'Price', find the total revenue generated by each product.

```python
# Sample DataFrame
data = {'Product':['Item1','Item2','Item3','Item4'],
        'Quantity':[10,5,7,22],
        'Price':[25,18,30,22]}
df = pd.DataFrame(data)

# Solution
revenue = df.groupby('Product')['Quantity','Price'].apply(lambda x: (x['Quantity']*x['Price']).sum())
revenue
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
df
```

```python
df = df.drop(['Revenue'], axis=1)
df
```

Given a DataFrame `df` with columns 'City', 'Year', and 'Temperature', find the average temperature for each city for each year.

```python
# Sample DataFrame
data = {'City':['NYC','LA','NYC','LA','LA'], 'Year':[2019,2020,2019,2020,2021],'Temperature':[21,22,19,23,29]}
df = pd.DataFrame(data)

# Solution
city_avg_temp = df.groupby(['City','Year'])['Temperature'].mean()
city_avg_temp
```
