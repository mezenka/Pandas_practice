### **Merge**  

MERGE is a more sophisticated, complicated. and flexible version of JOIN
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

```python
day01trips.merge(taxi_zones, left_on='pickup_location_id', right_on='zone_id', how='left')
```

```python
# how= can be {‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default is ‘inner’
# left = SQL LEFT OUTER JOIN
# right = SQL RIGHT OUTER JOIN
# outer = SQL FULL OUTER JOIN
# inner = SQL INNER JOIN
# cross - creates the cartesian product from both frames, preserves the order of the left keys
## cartesian product: all possible combinations of elements from the two original sets, denoted as A × B = {(1, a), (1, b), (2, a), (2, b)}, the resulting set contains all possible ordered pairs, named after René Descartes
# just for fun replacing the join direction to see how result would change
day01trips.merge(taxi_zones, left_on='pickup_location_id', right_on='zone_id', how='right')
# returns NULLs for unused pickup locations
```

```python
# how='outer'
day01trips.merge(taxi_zones, left_on='pickup_location_id', right_on='zone_id', how='right')
# returns more rows than in the left tables because of some unused zones, which appear with NULLs for the left table columns
```

```python
# inner
day01trips.merge(taxi_zones, left_on='pickup_location_id', right_on='zone_id', how='inner')
# resulting number or rows smaller than in the left table, because only the records with matching ON fields are returned
```

### **Join**  
used quite rarely, mostly when it is neccesary to join tables by their indexes  
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html

```python
day01trips.join(taxi_zones, how='left')
# without any details joins two tables using their indexes
```

```python
# adding the ON means JOIN will be lookign for specified column in the left table, but would still join it on the index of the right table
day01trips.join(taxi_zones, how='left', on='pickup_location_id')
```

```python
# workaround - specifically setting the fields to join on as indexes
day01trips.set_index('pickup_location_id').join(taxi_zones.set_index('zone_id'), how='left', on='pickup_location_id')
```

### **Concat**  
https://pandas.pydata.org/docs/reference/api/pandas.concat.html

```python
# for two or more dataframes
pd.concat([day01trips, day02trips, day03trips])
# adding end-to-end, just combining
# tables NEED to have the same columns with the same names
# indexes remain same as in source tables, so they start from 0 with every new table
```

```python
# to refresh indexes
pd.concat([day01trips, day02trips, day03trips], ignore_index=True)
```

```python
# for when it is necessarly to add extra columns to a dataframe
pd.concat([day01trips, day02trips, day03trips], axis=1)
```
